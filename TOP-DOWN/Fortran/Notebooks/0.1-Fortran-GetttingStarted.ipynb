<h1> Fortran as a Calculator </h1>

<p> programs written in early versions of FORTRAN language must follow special rules that were related to punch cards to store program codes. Fortran 77 being the last version where the character set included only uppercase letters. </p>
<H4>The initial release of FORTRAN for the IBM 704 contained 32 types of statements</H4>
<ul>
    <li>DIMENSION and EQUIVALENCE statements</li>
    <li>Assignment statements</li>
    <li>Three-way arithmetic IF statement</li>
    <li> Control statements for checking exceptions (IF ACCUMULATOR OVERFLOW, IF QUOTIENT OVERFLOW IF DIVIDE CHECK); and control statements for manipulating sense switches and sense lights (IF (SENSE SWITCH), IF (SENSE LIGHT), and SENSE LIGHT)</li>
    <li>GO TO, computed GO TO, ASSIGN and assigned GO TO</li>
    <li> DO loops</li>
    <li> Formatted I/O: FORMAT, READ, READ INPUT TAPE, WRITE OUTPUT TAPE, PRINT, and PUNCH</li>
    <li> Unformatted I/O: READ TAPE, READ DRUM, WRITE TAPE, and WRITE DRUM</li>
    <li> Other I/O: END FILE, REWIND, and BACKSPACE</li>
    <li>PAUSE, STOP, and CONTINUE</li>
    <li>FREQUENCY statement (for providing optimization hints to the compiler</li>
</ul>

<h5> Sample Code of an early version of FORTRAN</h5>
<pre>C---- THIS PROGRAM READS INPUT FROM THE CARD READER,
C---- 3 INTEGERS IN EACH CARD, CALCULATE AND OUTPUT
C---- THE SUM OF THEM.
  100 READ(5,10) I1, I2, I3
   10 FORMAT(3I5)
      IF (I1.EQ.0 .AND. I2.EQ.0 .AND. I3.EQ.0) GOTO 200
      ISUM = I1 + I2 + I3
      WRITE(6,20) I1, I2, I3, ISUM
   20 FORMAT(7HSUM OF , I5, 2H, , I5, 5H AND , I5,
     *   4H IS , I6)
      GOTO 100
  200 STOP
      END</pre>

<p>One code line is recorded on one punch card. A punch card has 80 columns; columns from 73 to 80 are used to record a sequence number of each card. Each card cannot record more than one statement. FORTRAN statement code must be written in columns 7 to 72. Columns 1 to 5 are used to enter numbers to identify statements to be referenced in GOTO, DO, READ, and WRITE statements. Column 6 is used to mark a card to continue a long statement from the previous card. A letter "C" in columns 1 indicates that this card contains a comment, not a FORTRAN statement.</p>

<h2>Evolution of FORTRAN</h2>
<p>In 1958 FORTRAN II supported procedural proramming with six new statements</p>
<li>SUBROUTINE, FUNCTION, and END</li>
<li>CALL and RETURN</li>
<li>COMMON</li>

<h3>Heron's formula</h3>

<pre>C AREA OF A TRIANGLE WITH A STANDARD SQUARE ROOT FUNCTION
C INPUT - TAPE READER UNIT 5, INTEGER INPUT
C OUTPUT - LINE PRINTER UNIT 6, REAL OUTPUT
C INPUT ERROR DISPLAY ERROR OUTPUT CODE 1 IN JOB CONTROL LISTING
      READ INPUT TAPE 5, 501, IA, IB, IC
  501 FORMAT (3I5)
C IA, IB, AND IC MAY NOT BE NEGATIVE OR ZERO
C FURTHERMORE, THE SUM OF TWO SIDES OF A TRIANGLE
C MUST BE GREATER THAN THE THIRD SIDE, SO WE CHECK FOR THAT, TOO
      IF (IA) 777, 777, 701
  701 IF (IB) 777, 777, 702
  702 IF (IC) 777, 777, 703
  703 IF (IA+IB-IC) 777, 777, 704
  704 IF (IA+IC-IB) 777, 777, 705
  705 IF (IB+IC-IA) 777, 777, 799
  777 STOP 1
C USING HERON'S FORMULA WE CALCULATE THE
C AREA OF THE TRIANGLE
  799 S = FLOATF (IA + IB + IC) / 2.0
      AREA = SQRTF( S * (S - FLOATF(IA)) * (S - FLOATF(IB)) *
     +     (S - FLOATF(IC)))
      WRITE OUTPUT TAPE 6, 601, IA, IB, IC, AREA
  601 FORMAT (4H A= ,I5,5H  B= ,I5,5H  C= ,I5,8H  AREA= ,F10.2,
     +        13H SQUARE UNITS)
      STOP
      END</pre>


<p> <b>FORTRAN III</b> allowed for inline assembly code, however never released as a product. Still included machine-dependent features, as well as Boolean expression support. <b>FORTRAN IV</b> removed the machine-dependent features of FORTRAN II while adding new features such as a LOGICAL data type, logical Boolean expressions, and the logical IF statement as an alternative to the arithmetic IF statement. Type declarations were added, along with an IMPLICIT statement to override earlier conventions that variables are INTEGER if their name begins with I, J, K, L, M, or N; and REAL otherwise. </p>

<h4>FORTRAN 66 became the first industry-standard version of FORTRAN</h4>
<p>FORTRAN 66 included:</p>
<ul>
    <li>Main program, SUBROUTINE, FUNCTION, and BLOCK DATA program units</li>
    <li>INTEGER, REAL, DOUBLE PRECISION, COMPLEX, and LOGICAL data type</li>
    <li>COMMON, DIMENSION, and EQUIVALENCE statements</li>
    <li>DATA statement for specifying initial values</li>
    <li>Intrinsic and EXTERNAL functions</li>
    <li>Assignment statement</li>
    <li>GO TO, computed GO TO, assigned GO TO and ASSIGN statements</li>
    <li>Logical IF and arithmetic (three-way) IF statements</li>
    <li>DO loop statements</li>
    <li>READ, WRITE, BACKSPACE, REWIND, and ENDFILE statements for sequential I/O</li>
    <li>FORMAT statement and assigned format</li>
    <li>CALL, RETURN, PAUSE, and STOP statements</li>
    <li>Hollerith constants in DATA and FORMAT statements, and as arguments to procedures</li>
    <li>Identifiers of up to six characters in length</li>
    <li>Comment lines</li>
    <li>END line</li>
</ul>
<h3>UPDATED HERON PROGRAM</h3>
<pre>C AREA OF A TRIANGLE WITH A STANDARD SQUARE ROOT FUNCTION
C INPUT - TAPE READER UNIT 5, INTEGER INPUT
C OUTPUT - LINE PRINTER UNIT 6, REAL OUTPUT
C INPUT ERROR DISPLAY ERROR OUTPUT CODE 1 IN JOB CONTROL LISTING
      READ (5, 501) IA, IB, IC
  501 FORMAT (3I5)
C
C IA, IB, AND IC MAY NOT BE NEGATIVE OR ZERO
C FURTHERMORE, THE SUM OF TWO SIDES OF A TRIANGLE
C MUST BE GREATER THAN THE THIRD SIDE, SO WE CHECK FOR THAT, TOO
      IF (IA .GT. 0 .AND. IB .GT. 0 .AND. IC .GT. 0) GOTO 10
        WRITE (6, 602)
  602   FORMAT (42H IA, IB, AND IC MUST BE GREATER THAN ZERO.)
        STOP 1
   10 CONTINUE
C
      IF (IA+IB-IC .GT. 0
     +    .AND. IA+IC-IB .GT. 0
     +    .AND. IB+IC-IA .GT. 0) GOTO 20
        WRITE (6, 603)
  603   FORMAT (50H SUM OF TWO SIDES MUST BE GREATER THAN THIRD SIDE.)
        STOP 1
   20 CONTINUE
C
C USING HERON'S FORMULA WE CALCULATE THE
C AREA OF THE TRIANGLE
      S = (IA + IB + IC) / 2.0
      AREA = SQRT ( S * (S - IA) * (S - IB) * (S - IC))
      WRITE (6, 601) IA, IB, IC, AREA
  601 FORMAT (4H A= ,I5,5H  B= ,I5,5H  C= ,I5,8H  AREA= ,F10.2,
     +        13H SQUARE UNITS)
      STOP
      END</pre>
