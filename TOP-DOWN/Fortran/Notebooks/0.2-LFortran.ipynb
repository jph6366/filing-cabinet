<p>Now lets write a program unit, the fundamental concept of a Fortran program.
A program unit is a a main program, an external subprogram, a module, a submodule, or a block data program unit.</p>

<b>A program unit consists of a set of nonoverlapping scoping units</b>

#### A scoping unit
BLOCK construct, derived-type definition, interface body, program unit, or subprogram, excluding all nested scoping units in it

##### The BLOCK construct is an executable construct that can contain declarations.

##### Actions on a variable local to a BLOCK construct do not affect any variable of the same name outside the construct. For example,


```fortran
integer :: x
x = 1
integer :: y
logical :: swapxy
IF (swapxy) THEN
    BLOCK
        REAL (KIND (x)) tmp
        tmp = x
        x = y
        y = tmp
    END BLOCK
END IF
print *, x
```
