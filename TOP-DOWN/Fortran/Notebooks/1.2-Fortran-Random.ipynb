<h1> RANDOM_NUMBER -- Pseudo-random number</h1>
<h2>Synopsis</h2>
<pre>CALL RANDOM_NUMBER(HARVEST)</pre>
<p> for REAL argument, returns a single pseudorandom number of an array of pseudorandom numbers from the uniform distribution over the range <pre>0<=x<1</pre></p>
<p> for UNSIGNED argument, returns a single pseudorandom number of an array of pseudorandom numbers over the range <pre>0<=x<=HUGE(HARVEST)</pre></p>

<p> the runtime library implements the xoshiro256** pseudorandom number generator (PRNG). This generator has a period <pre>2^{256} - 1</pre> and when using multiple threads up to <pre>2^{128}</pre> threads can each generate <pre>2^{128}</pre> random numbers before any aliasing occurs </p>
<p> In a multi-threaded program, each thread has its own random number state</p>

```fortran
program random_example
    implicit none
    real :: r(5,5)
    call random_seed()
    call random_number(r)
    print '(5(f10.6, 1x))', r
end program random_example
```
