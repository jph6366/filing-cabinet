## SQL: Structured Query Language
SQL is domain-specific language used to interact with database servers.
**pronunciation**: I prefer saying `squeal` to save time

- SQL is the most common standardized language used to access databases
    - SQL can be entered directly, embed statements into code (written in another programming language), or use language-specific API that hides SQL syntax
 
- SQL is defined by the ANSI/ISO SQL Standard
    - been evolving since 1986

### What is MySQL?
the most popular open source SQL management system, and support by Oracle
_MySQL is a database management_
- **a database** is a structured collection of data
    - to add, access, and process data stored in database

_MySQL databases are relational_
- **a relational database** stores data in separate tables rather than putting all data in one big store
    - the database structures are organized into physical files optimized for speed
    - the logical model with objects such as databases, tables, views, rows, and columns
    - set up rules governing the relationships between different data fields
        - one-to-one, one-to-many, unique, required or optional, and 'pointers' between different tables
        - these rules are enforced so the application never sees inconsistent, duplicate, orphaned, out-of-date, or missing data

_MySQL Server works in client/server or embedded systems_
- the MySQL database software is a client/server system that consists of a multithreaded SQL server that supports different back ends, several different client programs and libraries, administrative tools, and a wide range of application programming interfaces
- MySQL Server as an embedded multithreaded library that you can link into your application to get a smaller, faster, easier-to-manage standalone product.
